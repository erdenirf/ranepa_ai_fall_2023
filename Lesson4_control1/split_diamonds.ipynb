```python
import pandas as pd
from sklearn.model_selection import train_test_split 
```

```python
df = pd.read_csv("diamonds.csv", index_col=0)
```

```python
df
```

```python
X = df.drop('price',axis=1).values
y = df['price'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state = 42)
```

```python
df_train_x = pd.DataFrame(X_train, columns=df.drop('price',axis=1).columns)
```

```python
df_train_x
```

```python
df_train_y = pd.DataFrame(y_train, columns=['price'])
```

```python
df_train_y
```

```python
df_train = pd.concat([df_train_x, df_train_y], axis=1)
```

```python
df_train.to_csv("briliants_train.csv")
```

```python
df_test_y = pd.DataFrame(X_test, columns=df.drop('price',axis=1).columns)
```

```python
df_test_y.to_csv("briliants_test.csv")
```

```python
pd.DataFrame(y_test, columns=['price']).to_csv("briliants_y_true.csv")
```
